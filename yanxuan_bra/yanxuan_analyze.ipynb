```python
import jieba
```

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
wlist=''.join(df1['content'].tolist())
```

```python
from PIL import Image
```

```python
wl=jieba.analyse.extract_tags(wlist, topK=1000, withWeight=False, allowPOS=())
words=" ".join(wl)
PICTURE='head.jpg'
background = np.array(Image.open(PICTURE))
stopwords = set(STOPWORDS)
wc = WordCloud(background_color="white",
               font_path="msyh.ttf",
                mask=background,
                stopwords=stopwords)
wc.generate(words)
wc.to_file('bbb.png')
```

```python
#wl=jieba.analyse.extract_tags(wlist, topK=1000, withWeight=True, allowPOS=())
wl=jieba.lcut(wlist)
words=" ".join(wl)
PICTURE='head.jpg'
background = np.array(Image.open(PICTURE))
stopwords = set(STOPWORDS)
wc = WordCloud(background_color="white",
               font_path="msyh.ttf",
                mask=background,
                stopwords=stopwords)
wc.generate(words)
wc.to_file('bbb1.png')
```
